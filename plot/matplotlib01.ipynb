```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)

# 散布図の描画
plt.scatter(x, y)
plt.show()
```
