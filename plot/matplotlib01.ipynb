```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)

# 散布図の描画
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.choice(np.arange(100), 50)
y = np.random.choice(np.arange(100), 50)

# 散布図の描画
plt.scatter(x, y)
plt.show()
```

```python
days = np.arange(1, 11)
weight = np.array([10, 14, 18, 20, 18, 16, 17, 18, 20, 17])
# 表示の設定
plt.ylim([0, weight.max()+1])
plt.xlabel("days")
plt.ylabel("weight")

# 円マーカーを赤色でプロットし折れ線グラフ作成
plt.plot(days, weight, marker="o", markerfacecolor="r")
plt.show()
```

```python
plt.ylim([0, weight.max()+1])
plt.xlabel("days")
plt.ylabel("weight")

# 円マーカーを赤色でプロットし、青の破線の折れ線グラフ作成
plt.plot(days, weight, linestyle="--", color="b", marker="o", markerfacecolor="r")
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [12, 41, 32, 36, 21, 17]

# 棒グラフ
plt.bar(x, y)
plt.show()
```

```python
labels = ["Apple", "Orange", "Banana", "Pineapple", "Kiwifruit", "Strawberry"]

# 横軸にラベルを設定して棒グラフ作成
plt.bar(x, y, tick_label= labels)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y1 = [12, 41, 32, 36, 21, 17]
y2 = [43, 1, 6, 17, 17, 9]
labels = ["Apple", "Orange", "Banana", "Pineapple", "Kiwifruit", "Strawberry"]


# 横軸にラベルを設定して積み上げ棒グラフ作成
plt.bar(x, y1, tick_label=labels)
plt.bar(x, y2, bottom=y1)

# 凡例設定
plt.legend(("y1", "y2"))
plt.show()
```

```python
np.random.seed(0)
data = np.random.randn(10000)

# ヒストグラム作成
plt.hist(data)
plt.show()
```

```python
# ビン数20のヒストグラム
plt.hist(data, bins=20 )
plt.show()
# ビン数自動生成
plt.hist(data, bins="auto") 
plt.show()
```

```python
# 正規化されたビン数30のヒストグラム
plt.hist(data, bins=30, normed=True)
plt.show()

# 正規化されたビン数100の累積ヒストグラム
plt.hist(data, bins=100, normed=True, cumulative=True)
plt.show()
```

```python
np.random.seed(0)
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)

# マーカーの種類を四角、色を赤の散布図
plt.scatter(x, y, marker="s" , color="r" )
plt.show()
```

```python
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)
z = np.random.choice(np.arange(100), 100)

# zの値に応じて、マーカーの大きさが変えてプロット
plt.scatter(x, y, s=z)
plt.show()
```

```python
# zの値に応じて、マーカーの濃さを青系統に変えてプロット
plt.scatter(x, y, c=z , cmap="Blues")
plt.show()
```

```python
# zの値に応じて、マーカーの濃さを青系統に変えてプロット
plt.scatter(x, y, c=z, cmap="Blues")
# カラーバー表示
plt.colorbar()
plt.show()
```

```python
data = [60, 20, 10, 5, 3, 2]

# dataを円グラフとして可視化
plt.pie(data)
plt.show() #このままだと楕円グラフ

# 円グラフを円形にする
plt.axis("equal")
plt.pie(data)
plt.show()
```

```python
labels1 = ["Apple", "Orange", "Banana", "Pineapple", "Kiwifruit", "Strawberry"]

# dataにlabels1のラベルをつけた円グラフ
plt.pie(data, labels=labels1)
plt.axis("equal")
plt.show()
```

```python
explode1 = [0, 0, 0.1, 0, 0, 0]

# dataにlabels1のラベルをつけて、Bananaを目立たせた円グラフを描画
plt.pie(data, labels=labels1, explode=explode1)
plt.axis("equal")
plt.show()
```

```python
# 3D描画を行うために必要なライブラリ
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t, t)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Figureオブジェクト作成
fig = plt.figure(figsize=(6,6))
# サブプロット作成
ax = fig.add_subplot(1,1,1,projection="3d")

ax.plot_surface(X, Y, Z)
plt.show()
```

```python
x = y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2)/2) / (2*np.pi)

# Figureオブジェクト作成
fig = plt.figure(figsize=(6, 6))
# サブプロット作成
ax  = fig.add_subplot(1, 1, 1, projection="3d")
# 曲面表示
ax.plot_surface(X,Y,Z)

plt.show()
```

```python
# Figureオブジェクト作成
fig = plt.figure(figsize=(5, 5))
# サブプロット作成
ax1 = fig.add_subplot(111, projection="3d")

# x, y, zの位置を決める
xpos = [i for i in range(10)]
ypos = [i for i in range(10)]
zpos = np.zeros(10)
   
# x, y, zの増加量を決める
dx = np.ones(10)
dy = np.ones(10)
dz = [i for i in range(10)]

# 3次元bar作成
ax1.bar3d(xpos, ypos, zpos, dx, dy, dz)
plt.show()
```

```python
X = np.random.randn(1000)
Y = np.random.randn(1000)
Z = np.random.randn(1000)

# Figureオブジェクト作成
fig = plt.figure(figsize=(6, 6))

# サブプロット作成
ax2  = fig.add_subplot(1, 1, 1, projection="3d")

# X,Y,Zを1次元変換
x = np.ravel(X)
y = np.ravel(Y)
z = np.ravel(Z)

# 3D散布図作成
ax2.scatter3D(x, y, z)
plt.show()
```

```python
# カラーマップを表示するためのライブラリ
from matplotlib import cm

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t, t)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Figureオブジェクト作成
fig = plt.figure(figsize=(6, 6))

# サブプロット作成
ax3  = fig.add_subplot(1,1,1, projection="3d")

# zにカラーマップ適用
ax3.plot_surface(X, Y, Z, cmap=cm.coolwarm )
plt.show()
```
